```python
A=22;B=81
```

```python
def dopCode(n):
    return 2**len_o-n
```

```python
def bin_sum(a,b):
    a = list(map(int,list(a)))
    b = list(map(int,list(b)))
    d = len(a)-len(b)
    if d > 0:
        b = [0]*d+b
    if d < 0:
        a = [0]*abs(d)+a
    c=0
    s=''
    for i,j in zip(a[::-1],b[::-1]):
        s = s + chr((i+j+c)%2+ord('0'))
        c = (i+j+c)//2
    if c == 1:
        s = s + '1'
    s=s[::-1]
    return s
```

1. Умножение в дополнительных кодах с применением коррекции.

```python
def mul_with_correction(A,B):
    if A < 0:
        A=dopCode(-A)
    binA = bin(A)[2:]
    binA = '0'*(len_o-len(binA))+binA
    
    if B < 0:
        B=dopCode(-B)
    binB = bin(B)[2:]
    binB = '0'*(len_o-len(binB))+binB
    
    df = pd.DataFrame([],columns=['Операция и действие','СЧП(старшие разряды)',
                              'Множитель и СЧП(младшие разряды)','Пояснение'])
    a = list(map(int,list(binA)))
    b = list(map(int,list(binB)))
    s = '0'*(len_o*2)
    for j,i in enumerate(b[::-1]):
        k = ['','','','']
        if i == 1:
            k[0]+='A'+('доп' if a[0] == '1' else 'пр')+'\nСЧП\n'
        k[0] +='СЧП->'
        if i == 1:
            k[1] += binA[:len_o]+'\n'
            s = bin_sum(s,a+[0]*len_o)
            if len(s)!=len_o*2+1:
                k[1] += s[:len_o]+'\n'
            else:
                k[1] += s[1:len_o+1]+'\n'
            k[3] = 'Сложение СЧП с множимым\n'
        k[3]+='Модифицированный сдвиг СЧП и множителя вправо'
        if i == 1:
            if len(s) == len_o*2:
                k[2] += (('\n'+s[len_o:len_o+j]+'|') if j > 0 else '')+binB[:(-j if j > 0 else None)]+'\n'
            else:
                k[2] += '\n'+s[len_o+1:len_o+j+1]+'|'+binB[:(-j if j > 0 else None)]+'\n'
        if len(s)==len_o*2+1:
            s = s[:len_o*2]
        elif s[0]=='0':
            s = '0'+s[:len_o*2-1]
        else:
            s = '1'+s[:len_o*2-1]
        k[2] += s[len_o:len_o+j+1]+(('|'+binB[:-j-1]) if j!=len(b)-1 else '')
        k[1] += s[:len_o]

        k = pd.DataFrame([k],columns=['Операция и действие','СЧП(старшие разряды)',
                                  'Множитель и СЧП(младшие разряды)','Пояснение'])
        df = pd.concat((df,k),axis=0,ignore_index=True)
    df = pd.concat((pd.DataFrame([["СЧП","0"*len_o,binB,"Обнуление старших разрядов СЧП"]],columns=['Операция и действие','СЧП(старшие разряды)',
                                  'Множитель и СЧП(младшие разряды)','Пояснение']),df),axis=0,ignore_index=True)
    if (b[0] == 1):
        binA = bin(dopCode(A))[2:]
        binA = '0'*(len_o-len(binA))+binA
        a = list(map(int,list(binA)))
        k = ['','','','']
        if i == 1:
            k[0]+='-[A]доп\nСЧП\n'
        k[1] += binA+'\n'
        s = bin_sum(s,a+[0]*len_o)
        if len(s)!=len_o*2+1:
            k[1] += s[:len_o]+'\n'
        else:
            k[1] += s[1:len_o+1]+'\n'
        k[3] = 'Коррекция результата сложение старших разрядов СЧП с дополнением множимого'
        k[2] = '\n'
        if len(s) != len_o*2+1:
            k[2] = s[len_o:]
        else:
            k[2] = s[len_o+1:]

        k = pd.DataFrame([k],columns=['Операция и действие','СЧП(старшие разряды)',
                                  'Множитель и СЧП(младшие разряды)','Пояснение'])
        df = pd.concat((df,k),axis=0,ignore_index=True)
    return df
```

2. Умножение в дополнительных кодах без коррекции.

```python
def mul_without_correction(A,B):
    if A < 0:
        A=dopCode(-A)
    binA = bin(A)[2:]
    binA = '0'*(len_o-len(binA))+binA
    
    binAdop = bin(dopCode(A))[2:]
    binAdop = '0'*(len_o-len(binAdop))+binAdop
    
    if B < 0:
        B=dopCode(-B)
    binB = bin(B)[2:]
    binB = '0'*(len_o-len(binB))+binB
    
    df = pd.DataFrame([],columns=['Операция и действие','СЧП(старшие разряды)',
                              'Множитель и СЧП(младшие разряды)','Пояснение'])
    a = list(map(int,list(binA)))
    adop = list(map(int,list(binAdop)))
    b = list(map(int,list(binB)))
    s = '0'*(len_o*2)
    prev = None
    for j,i in enumerate(b[::-1]):
        action=(i != prev)
        k = ['','','','']
        if i == 1 and i != prev:
            k[0]+='-[A]'+('пр' if a[0] == '1' else 'доп')+'\nСЧП\n'
            k[1] += binAdop[:len_o]+'\n'
            s = bin_sum(s,adop+[0]*len_o)
            if len(s)!=len_o*2+1:
                k[1] += s[:len_o]+'\n'
            else:
                k[1] += s[1:len_o+1]+'\n'
            k[3] = 'Вычитание множимого из СЧП\n'
        if i == 0 and i != prev:
            k[0]+='[A]'+('доп' if a[0] == '1' else 'пр')+'\nСЧП\n'
            k[1] += binA[:len_o]+'\n'
            s = bin_sum(s,a+[0]*len_o)
            if len(s)!=len_o*2+1:
                k[1] += s[:len_o]+'\n'
            else:
                k[1] += s[1:len_o+1]+'\n'
            k[3] = 'Сложение СЧП с множимым\n'
            
        k[0] +='СЧП->'
        k[3]+='Модифицированный сдвиг СЧП и множителя вправо'
        if action:
            if len(s) == len_o*2:
                k[2] += (('\n'+s[len_o:len_o+j]+'|') if j > 0 else '')+binB[:(-j if j > 0 else None)]+'\n'
            else:
                k[2] += '\n'+s[len_o+1:len_o+j+1]+'|'+binB[:(-j if j > 0 else None)]+'\n'
        if len(s)==len_o*2+1:
            s = s[1:len_o*2]
        if s[0]=='0':
            s = '0'+s[:len_o*2-1]
        else:
            s = '1'+s[:len_o*2-1]
        k[2] += s[len_o:len_o+j+1]+(('|'+binB[:-j-1]) if j!=len(b)-1 else '')
        k[1] += s[:len_o]

        k = pd.DataFrame([k],columns=['Операция и действие','СЧП(старшие разряды)',
                                  'Множитель и СЧП(младшие разряды)','Пояснение'])
        df = pd.concat((df,k),axis=0,ignore_index=True)
        prev = i
    df = pd.concat((pd.DataFrame([["СЧП","0"*len_o,binB,"Обнуление старших разрядов СЧП"]],columns=['Операция и действие','СЧП(старшие разряды)',
                                  'Множитель и СЧП(младшие разряды)','Пояснение']),df),axis=0,ignore_index=True)
    return df
```
